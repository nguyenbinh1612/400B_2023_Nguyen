```python
# import modules
import numpy as np
import astropy.units as u
import astropy.table as tbl

# import plotting modules
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

from ReadFile import Read
from CenterOfMass_Solution import CenterOfMass
```

```python
class MassProfile:
# Class to determine the mass distribution of a given galaxy 
# at a simulation snapshot

    def __init__(self, galaxy, snap):
        ''' # Class to determine the mass distribution of a given galaxy 
        at a simulation snapshot
            
            PARAMETERS
            ----------
            galaxy: string
                a string with galaxy name, e.g. “MW”, “M31” or “M33”
            snap: int
                snapshot number, e.g. 0, 1, etc. 
        '''
        
        # add a string of the filenumber to the value “000”
        ilbl = "000" + str(snap)
        # remove all but the last 3 digits
        ilbl = ilbl[-3:]
        self.filename=str(galaxy) + '_' + ilbl + ".txt"
        
        # read data in the given file using Read
        self.time, self.total, self.data = Read(self.filename) 
        
        # read the mass, positions of the relevant galaxy at the relevant snapshot
        self.m = self.data['m']*10**10 # unit not assigned just yet
        self.x = self.data['x']*u.kpc
        self.y = self.data['y']*u.kpc
        self.z = self.data['z']*u.kpc
        
        # save the galaxy's name for later use
        self.gname = galaxy
        
    def MassEnclosed(self, particle_type, radius_array):
        ''' # This function calculates how much of a certain particle type in a galaxy
        is enclosed within each radius on a given radius array
            
            INPUTS
            ----------
            particle_type: int; 1, 2, or 3
                particle type to use for COM calculations
                note to self: 1 is for DM halo, 2 for disk, 3 for bulge. 
            radius_array: NumPy array
                NumPy array of radii to enclose galaxy mass (in kpc)
                
            OUTPUTS
            ----------
            mass_enclosed: NumPy array
                the mass of a particular particle type enclosed by each radius in the radius array
            
        '''
        
        # first, find the center of mass of the system using the CenterOfMass code
        galaxyCOM = CenterOfMass(self.filename, particle_type)
        
        p_COM = galaxyCOM.COM_P(0.1)
        
        # create an array to store indexes of particles of desired particle_type                                
        self.index = np.where(self.data['type'] == particle_type)
        
        # read the mass, positions of the relevant galaxy at the relevant snapshot
        m = self.m[self.index]
        x = self.x[self.index]
        y = self.y[self.index]
        z = self.z[self.index]

        # find the magnitude of the position vector of all the galaxy's particles
        x_new = x - p_COM[0]
        y_new = y - p_COM[1]
        z_new = z - p_COM[2]
        p_mag = np.sqrt(x_new**2 + y_new**2 + z_new**2)
        
        # create an empty array that's as long as the radius array input in order to find the mass 
        # enclosed at each radius, which allows us to construct a mass profile
        mass_enclosed = np.zeros(np.size(radius_array))
        
        # loop over all the radii in the radius array to find the mass enclosed within each of them
        for idx in range(np.size(radius_array)):   
            # *u.kpc because I'm assuming the array wouldn't be given in astropy units
            mass_enclosed[idx] = np.sum(m[p_mag < radius_array[idx]*u.kpc])
        return mass_enclosed*u.Msun
    
    def MassEnclosedTotal(self, radius_array):
        ''' # This function calculates the total enclosed mass of the galaxy with respect 
        to radius. Note that we have to make an exception for M33, which lacks a bulge.
            
            INPUTS
            ----------
            radius_array: NumPy array
                NumPy array of radii to enclose galaxy mass (in kpc)
                
            OUTPUTS
            ----------
            TotalMassEnclosed: NumPy array
                the total mass of all particle types enclosed by each radius in the radius array
            
        '''
        
        HaloMassEnclosed = self.MassEnclosed(1, radius_array)
        DiskMassEnclosed = self.MassEnclosed(2, radius_array)
        
        if self.gname == "M33":
            TotalMassEnclosed = HaloMassEnclosed + DiskMassEnclosed 
        else:
            BulgeMassEnclosed = self.MassEnclosed(3, radius_array)
            TotalMassEnclosed = HaloMassEnclosed + DiskMassEnclosed + BulgeMassEnclosed
        
        return TotalMassEnclosed
    
    def HernquistMass(self, r, a, Mhalo):
        ''' # This function calculates the Hernquist mass of a galaxy
        at a certain radius given the halo mass enclosed.
            
            INPUTS
            ----------
            r: astropy quantity
                the radius that encloses the halo mass (in kpc)
            a: astropy quantity
                a scaling factor (in kpc)
            Mhalo: astropy quantity
                the halo mass enclosed by the abovementioned radius (in Msun)
            
            OUTPUTS
            ----------
            M_Hernquist: astropy quantity
                the Hernquist mass of the galaxy, which is given by Mhalo*(r**2) / (a+r)**2
            
        '''
        
        M_Hernquist = Mhalo*((r*u.kpc)**2) / ((a+r)*u.kpc)**2
        return M_Hernquist
    
    def CircularVelocity(self, particle_type, radius_array):
        ''' # This function calculates the circular velocity of a galaxy based on
        the particle type's mass enclosed at particular radii.
            
            INPUTS
            ----------
            particle_type: int; 1, 2, or 3
                particle type to use for COM calculations
                note to self: 1 is for DM halo, 2 for disk, 3 for bulge. 
            radius_array: NumPy array
                NumPy array of radii to enclose galaxy mass (in kpc)
                
            OUTPUTS
            ----------
            vcirc: NumPy array
                the circular velocity of the galaxy, which is given as vcirc = (G*M(r)/r)**(1/2)
            
        '''
        
        M = self.MassEnclosed(particle_type, radius_array)
        G = 4.300917270036279e-06*(u.kilometer**2)*(u.kpc)/((u.second**2)*(u.Msun))
        vcirc = np.sqrt(G*M/radius_array)
        return vcirc
    
    def CircularVelocityTotal(self, radius_array):
        ''' # This function calculates the circular velocity of a galaxy based on
        ALL the particle mass enclosed at particular radii.
            
            INPUTS
            ----------
            radius_array: NumPy array
                NumPy array of radii to enclose galaxy mass (in kpc)
                
            OUTPUTS
            ----------
            vcirc_tot: NumPy array
                the circular velocity of the galaxy, which is given as vcirc = (G*M(r)/r)**(1/2)
            
        '''
        
        M = self.MassEnclosedTotal(radius_array)        
        G = 4.300917270036279e-06*(u.kilometer**2)*(u.kpc)/((u.second**2)*(u.Msun))
        vcirc_tot = np.sqrt(G*M/radius_array)
        return vcirc_tot
    
    def HernquistVCirc(self, r, a, Mhalo):
        ''' # This function calculates the Hernquist circular velocity of a galaxy
        at a certain radius given the Hernquist mass.
            
            INPUTS
            ----------
            r: astropy quantity
                the radius that encloses the halo mass (in kpc)
            a: astropy quantity
                a scaling factor (in kpc)
            Mhalo: astropy quantity
                the halo mass enclosed by the abovementioned radius (in Msun)
            
            OUTPUTS
            ----------
            vcirc_Hernquist: astropy quantity
                the Hernquist mass of the galaxy, which is given by Mhalo*(r**2) / (a+r)**2
            
        '''
        
        M_Hernquist = self.HernquistMass(r, a, Mhalo)
        G = 4.300917270036279e-06*(u.kilometer**2)*(u.kpc)/((u.second**2)*(u.Msun))
        vcirc_Hernquist = np.sqrt(G*M_Hernquist/(r*u.kpc))
        return vcirc_Hernquist
```

```python
# Now, we define a radius array
radius_array = np.arange(0.1, 30, 1.5)
```

```python
# Now, we create a plot of the mass profile (mass enclosed as a function of radius) 
# of each component of the MW to a radius of 30 kpc

# First, we call an object
MW = MassProfile("MW", 0)

# Next, we calculate how much mass is enclosed within 30 kpc per particle type
MWHaloMass = MW.MassEnclosed(1, radius_array)
MWDiskMass = MW.MassEnclosed(2, radius_array)
MWBulgeMass = MW.MassEnclosed(3, radius_array)

# Next, we calculate the total mass enclosed within 30 kpc
MWTotalMass = MW.MassEnclosedTotal(radius_array)

# Next, we compute the Hernquist mass
scale_factor = 0.01 # we try different scale factors until the Hernquist mass's plot matches the halo mass plot
MWHernquistMass = MW.HernquistMass(radius_array, scale_factor, MWHaloMass)
```

```python
plt.figure(figsize=(10,10))

plt.semilogy(radius_array, MWHaloMass, linewidth=2, label='halo mass')
plt.semilogy(radius_array, MWDiskMass, linewidth=2, label='disk mass')
plt.semilogy(radius_array, MWBulgeMass, linewidth=2, label='bulge mass')
plt.semilogy(radius_array, MWTotalMass, linewidth=2, label='total mass', color='black')
plt.semilogy(radius_array, MWHernquistMass, linewidth=2, \
             label='Hernquist mass, a = ' + str(scale_factor) + ' kpc', color='red', linestyle='-.')

plt.xlabel('distance from COM (kpc)', fontsize=17) 
plt.ylabel('mass enclosed (M$_\odot$)', fontsize=17)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(loc='lower right', prop={'size': 15})
plt.title('Mass profiles of the Milky Way at Snapshot 0', loc='center', fontsize=23)

plt.show()
```
